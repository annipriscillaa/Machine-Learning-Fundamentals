```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path="drive/MyDrive/ML"
x = pd.read_csv(path+'/test.csv')
```

```python
x.head()
```

```python
x_data = x['X2 house age']
y_data = x['Y house price of unit area']
```

```python
plt.scatter(x_data, y_data)
```

```python
import statistics as stc
```

```python
x_mean = stc.mean(x_data)
y_mean = stc.mean(y_data)
```

```python
B1_num = ((x_data - x_mean) * (y_data - y_mean)).sum()
B1_den = ((x_data - x_mean)**2).sum()
B1 = B1_num / B1_den
B0 = y_mean - (B1 * x_mean)
```

```python
plt.scatter(x_data,y_data)
plt.plot(x_data, B0 + B1*x_data, c = 'r')
```
